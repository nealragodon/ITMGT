```python
import requests
import json
import time

interval = 10
count = 100

info =[]
for i in range(count):
    page = requests.get("http://api.open-notify.org/iss-now.json")
    time.sleep(interval)
    
    info.append(page.json())

info
```

```python
latitudes =[]
for i in range(len(info)):
    latitudes.append(info[i]["iss_position"]["latitude"])

latitudes_list =[float(i) for i in latitudes]

latitudes_list
```

```python
longitudes=[]
for i in range(len(info)):
    longitudes.append(info[i]["iss_position"]["longitude"])
                      
longitudes_list =[float(i) for i in longitudes]
                      
longitudes_list
```

```python
timestamps= []
for i in range(len(info)):
    timestamps.append(info[i]["timestamp"])
    
timestamps
```

```python
import matplotlib.pyplot as plt

plt.scatter(latitudes_list, longitudes_list)
plt.title("Location of the International Space Station (ISS)")
plt.xlabel("latitude")
plt.ylabel("longitude")
```
